```python
from pyeoskit import eosapi
from pyeoskit import db
from pyeoskit import hello
_eosapi = hello._eosapi
db.reset()
args = {"from": 'eosio',
        "to": 'hello',
        "quantity": '0.0001 EOS',
        "memo": 'hello,world'
}
abi = eosapi.get_abi('eosio.token')
print(abi)
print(type(abi))
_eosapi.pack_args(abi, 'transfer', args)

r = eosapi.pack_args('eosio.token', 'transfer', args)
print(r)
eosapi.unpack_args('eosio.token', 'transfer', r)
```

```python
from pyeoskit import eosapi
args = {"from": 'eosio',
        "to": 'hello',
        "quantity": '0.0001 EOS',
        "memo": 'hello,world'
}
action = ['eosio.token', 'transfer', args, {'eosio':'active'}]
reference_block_id = eosapi.get_info().last_irreversible_block_id
r = eosapi.gen_transaction([action], 60*60, reference_block_id)
print(r)
```

```python
from pyeoskit import db
from pyeoskit import eosapi
args = {"from": 'eosio',
        "to": 'hello',
        "quantity": '0.0001 EOS',
        "memo": 'hello,world'
}

r = eosapi.pack_args('eosio.token', 'transfer', args)
assert r

args = {"from": 'hello',
        "to": 'eosio',
        "quantity": '0.0001 EOS',
        "memo": 'hello,world'
}
action = ['eosio.token', 'transfer', args, {'hello':'active'}]
reference_block_id = eosapi.get_info().last_irreversible_block_id
trx = eosapi.gen_transaction([action], 60, reference_block_id)
print(trx)

info = eosapi.get_info()
trx = eosapi.sign_transaction(trx, '5JEcwbckBCdmji5j8ZoMHLEUS8TqQiqBG1DRx1X9DN124GUok9s', info.chain_id)
print(trx)

trx = eosapi.pack_transaction(trx, 0)
print(trx)
eosapi.push_transaction(trx)
```

```python
from pyeoskit import eosapi
from pyeoskit import wallet

import os
if os.path.exists('mywallet.wallet'):
    os.remove('mywallet.wallet')
psw = wallet.create('mywallet')
wallet.unlock('mywallet', psw)
wallet.import_key('mywallet', '5JEcwbckBCdmji5j8ZoMHLEUS8TqQiqBG1DRx1X9DN124GUok9s')

args = {"from": 'hello',
        "to": 'eosio',
        "quantity": '0.0001 EOS',
        "memo": 'hello,world'
}
action = ['eosio.token', 'transfer', args, {'hello':'active'}]
eosapi.push_action(*action)
```
